# Creating a Graph using an Adjacency Matrix

An Adjacency Matrix is a 2D grid (a table) where the rows and columns represent locations. The numbers inside the grid represent the distance of the road connecting them. If the number is 0, it means there is no direct road.

````python
# Let's create a map with 4 locations: 0, 1, 2, and 3
# Rows and columns represent: [Node 0, Node 1, Node 2, Node 3]

adjacency_matrix = [
    [0, 5, 2, 0],  # Roads starting from Node 0
    [5, 0, 0, 4],  # Roads starting from Node 1
    [2, 0, 0, 7],  # Roads starting from Node 2
    [0, 4, 7, 0]   # Roads starting from Node 3
]

# How to read this matrix:
print(f"Distance from Node 0 to Node 1: {adjacency_matrix[0][1]}")
print(f"Distance from Node 0 to Node 3 (no direct road): {adjacency_matrix[0][3]}")
````

# Reading a graph from File
Reading a graph from a CSV file is a very common task in data analysis. Instead of typing everything into the terminal, you can just save your roads in an Excel or CSV file and let Python read it automatically.

**Step 1: Create Your CSV File**

First, create a simple spreadsheet in Excel or any text editor and save it as routes.csv.
Your file should look like this (with three columns: Start, End, and Distance):



````python
%%writefile routes.csv
Start,End,Distance
0,1,5
0,2,2
1,3,4
2,3,7
````

**The Python Code**

We will use Python's built-in csv module so you don't have to install any extra packages. Make sure your Python file and your routes.csv file are saved in the same folder.

````python
import csv

# 1. Open and read the CSV file to find out how many nodes we have
filename = "routes.csv"

all_edges = []
max_node_index = 0

with open(filename, mode='r') as file:
    # csv.DictReader automatically recognizes the header row (Start, End, Distance)
    csv_reader = csv.DictReader(file)

    for row in csv_reader:
        # Convert text from CSV into numbers
        start = int(row['Start'])
        end = int(row['End'])
        distance = int(row['Distance'])

        # Save this road for later
        all_edges.append((start, end, distance))

        # Keep track of the highest node number to know how big our grid needs to be
        if start > max_node_index:
            max_node_index = start

        if end > max_node_index:
            max_node_index = end

# The total number of nodes will be the highest index + 1 (since we start counting at 0)
num_nodes = max_node_index + 1
print(f'Total Node: ', num_nodes)


# 2. Create a blank grid (matrix) filled with zeros
adjacency_matrix = []
for i in range(num_nodes):
    row = [0] * num_nodes
    adjacency_matrix.append(row)


# 3. Fill the matrix using the data we read from the CSV
for start, end, distance in all_edges:
    # Assuming two-way roads:
    adjacency_matrix[start][end] = distance
    adjacency_matrix[end][start] = distance

print("Successfully created adjacency_matrix ")
````

````python

# 4. Print the final matrix beautifully
print(f" Successfully read {len(all_edges)} roads from '{filename}'")
print(f"Detected {num_nodes} total locations (Nodes 0 to {max_node_index})\n")
print("=== GENERATED ADJACENCY MATRIX ===")

for row in adjacency_matrix:
    print(row)
````

# Graph Metrics (Basic Network Properties)
In data analysis, we often want to look at "metrics" (measurements) of our network, like the Degree (how many direct connections a location has).

````python
# Using the matrix from above
def calculate_node_degrees(matrix):
    print("--- Graph Metrics: Node Degrees ---")
    for node_index in range(len(matrix)):
        connections = 0
        for road in matrix[node_index]:
            if road > 0:  # If a road exists (greater than 0)
                connections += 1
        print(f"Node {node_index} has {connections} direct connections.")

calculate_node_degrees(adjacency_matrix)
````

## Understanding Graph Metrics: Node Degrees

This function calculates a fundamental metric in network analysis called the **Degree** of a node. In plain terms, it counts **how many direct roads lead out of each location**.

It does this by analyzing the rows of your adjacency matrix grid. Here is how the logic works:

### 1. Setting up the Loop
* **`len(matrix)`**: Since the matrix is a square grid, its length tells us how many locations (nodes) exist in total. If it's a $4 \times 4$ grid, it loops 4 times (for Node 0, Node 1, Node 2, and Node 3).
* **`connections = 0`**: This sets up a temporary counter bucket for the current node. Every time we find a valid road, we will add 1 to this bucket.

### 2. Scanning the Grid Row
This is where the actual counting happens. The code looks at the specific row belonging to the node it is currently checking. For example, when checking **Node 0**, it looks at its row: `[0, 5, 2, 0]`.
* It checks each number (`road`) in that row one by one.
* **`if road > 0`**: Remember, a `0` means there is no road. Any number greater than 0 means a road exists.
* In Node 0's row, it sees `5` and `2`. Because both are greater than 0, the computer triggers `connections += 1` twice.

### 3. Printing the Metric
Once the code finishes checking all columns for that specific row, it prints out the final score.

### Visual Example of the Logic:
If we pass our matrix into this function, the math looks like this behind the scenes:
* **Row 0:** `[0, 5, 2, 0]` $\rightarrow$ Found two numbers $> 0$ ($5$ and $2$). **Score = 2**
* **Row 1:** `[5, 0, 0, 4]` $\rightarrow$ Found two numbers $> 0$ ($5$ and $4$). **Score = 2**
* **Row 2:** `[2, 0, 0, 7]` $\rightarrow$ Found two numbers $> 0$ ($2$ and $7$). **Score = 2**
* **Row 3:** `[0, 4, 7, 0]` $\rightarrow$ Found two numbers $> 0$ ($4$ and $7$). **Score = 2**

It then cleanly outputs these totals to the screen so you can quickly see which locations are the busiest hubs on your map!

# Graph Traversal (Breadth-First Search)

Traversal means "visiting every single room/location in the network" to make sure we don't miss any. Breadth-First Search (BFS) explores like a ripple in a pond: it checks all immediate neighbors first before moving further out.

````python
def traverse_graph_bfs(matrix, start_node):
    visited = []       # List to keep track of locations we've already seen
    queue = [start_node] # Our "to-do" list of locations to explore next

    while queue:
        # Take the first location off our to-do list
        current = queue.pop(0)

        if current not in visited:
            visited.append(current) # Mark it as visited

            # Find all neighbors of the current location
            for neighbor, road in enumerate(matrix[current]):
                if road > 0 and neighbor not in visited:
                    queue.append(neighbor) # Add neighbors to our to-do list

    return visited

# Let's see the order in which BFS visits the nodes
visit_order = traverse_graph_bfs(adjacency_matrix, start_node=0)
print(f"BFS Traversal Order: {visit_order}")
````

## Understanding Graph Traversal: Breadth-First Search (BFS)

Imagine you are standing at a starting location and want to explore an entire map of connected cities. **Breadth-First Search (BFS)** explores like a ripple in a pond: it checks all of your *immediate neighbors* first, before moving out to cities further away.

Think of it as exploring a map layer-by-layer, or level-by-level.



### 1. Setting Up the Lists
* **`visited = []`**: This is our "History Log". Once we have officially arrived at and explored a city, we add it here so we don't accidentally circle back and visit it again.
* **`queue = [start_node]`**: This is our "To-Do List". It keeps track of which cities we need to explore next. Because it is a *Queue*, it works on a **First-In, First-Out** basis (just like a line at a grocery store—the first person in line gets served first).

### 2. The Exploration Loop (`while queue:`)
The computer keeps running this loop as long as there are still cities left on our "To-Do List" (`queue`).

* **`current = queue.pop(0)`**: The computer grabs the very first city from the front of the line to explore it.
* **`if current not in visited:`**: It checks our History Log. If we haven't explored this city yet, it immediately logs it by doing `visited.append(current)`.

# Dijkstra's Shortest Path Algorithm
Dijkstra's algorithm is like Google Maps. It finds the absolute smartest, shortest route from your start location to every other location on the map, taking into account road distances.

````python
def dijkstra_shortest_path(matrix, start_node):
    num_nodes = len(matrix)
#    unknown_number = float('inf')
    unknown_number = 10000

    # Start by assuming all distances are unknown/infinity
    distances = [unknown_number] * num_nodes
    distances[start_node] = 0 # Distance to your starting point is 0

    visited = [False] * num_nodes

    for _ in range(num_nodes):
        # Find the node with the shortest known distance that we haven't visited yet
        min_dist = unknown_number
        current_node = -1

        for node in range(num_nodes):
            if not visited[node] and distances[node] < min_dist:
                min_dist = distances[node]
                current_node = node

        # Mark this node as finished
        visited[current_node] = True

        # Look at all neighbors of this node and update their distances
        for neighbor, road_length in enumerate(matrix[current_node]):
            if road_length > 0 and not visited[neighbor]:
                new_distance = distances[current_node] + road_length

                # If this new path is a shortcut, save it!
                if new_distance < distances[neighbor]:
                    distances[neighbor] = new_distance

    return distances

# Calculate shortest paths starting from Node 0
shortest_routes = dijkstra_shortest_path(adjacency_matrix, start_node=0)

print("--- Shortest Travel Distances From Node 0 ---")
for node, distance in enumerate(shortest_routes):
    print(f"Shortest distance to Node {node}: {distance} miles/km")
````
